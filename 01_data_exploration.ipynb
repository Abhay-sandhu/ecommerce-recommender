```python
import pandas as pd

# --- Load core tables ---
orders = pd.read_csv("drive/MyDrive/dataset/orders.csv")
order_products_prior = pd.read_csv("drive/MyDrive/dataset/order_products__prior.csv")
order_products_train = pd.read_csv("drive/MyDrive/dataset/order_products__train.csv")
products = pd.read_csv("drive/MyDrive/dataset/products.csv")
aisles = pd.read_csv("drive/MyDrive/dataset/aisles.csv")
departments = pd.read_csv("drive/MyDrive/dataset/departments.csv")

print("Orders:", orders.shape)
print("Prior order products:", order_products_prior.shape)
print("Train order products:", order_products_train.shape)
print("Products:", products.shape)

# --- Combine prior + train (full interaction log) ---
order_products = pd.concat([order_products_prior, order_products_train], axis=0)
print("Combined order_products:", order_products.shape)

# --- Merge with orders to get user_id ---
order_products = order_products.merge(orders[["order_id", "user_id"]], on="order_id", how="left")

print("Order_products merged:", order_products.shape)
order_products.head()

# --- Sampling (100k orders for fast prototyping) ---
sampled_order_ids = orders["order_id"].drop_duplicates().sample(n=100000, random_state=42)
order_products_sampled = order_products[order_products["order_id"].isin(sampled_order_ids)]

print("Sampled order_products:", order_products_sampled.shape)

# --- Merge product metadata ---
order_products_sampled = order_products_sampled.merge(products, on="product_id", how="left")
order_products_sampled = order_products_sampled.merge(aisles, on="aisle_id", how="left")
order_products_sampled = order_products_sampled.merge(departments, on="department_id", how="left")

print("Final sampled dataset:", order_products_sampled.shape)
order_products_sampled.head()
```
